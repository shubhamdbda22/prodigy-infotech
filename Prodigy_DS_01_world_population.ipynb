```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Here we imported data from our local disk and stored in df variable
df = pd.read_csv(r"C:/Users/hp/OneDrive/Desktop/CDAC-DBDA/2024/prodigy_infotech/task1/API_SP.POP.TOTL_DS2_en_csv_v2_45183.csv")
```

```python
df.head()
```

```python
## Here we droped unecessary columns
df = df.drop(df.index[:3]).reset_index(drop=True)
df.columns =df.iloc[0]
df = df.drop(df.index[:1]).reset_index(drop=True)
     
```

```python
df
```

```python
df.drop(['Indicator Name', 'Indicator Code', 'Country Code'], axis = 1, inplace=True)
df.head()
     
```

```python
# Select data for world population (assuming "World" is the country name)
world_data = df[df['Country Name'] == 'World']
```

```python

world_data
```

```python
world_data_df = world_data.melt(id_vars='Country Name', var_name='Year', value_name='Population')

# df.melt is the pandas function for unpivoting columns.
# id_vars='Country Name': This argument specifies the column to retain in each row of the unpivoted DataFrame. Here, we're keeping the "Country Name" column.
# var_name='Year': This argument renames the column containing the original column names (years in this case) to "Year" in the unpivoted DataFrame.
# value_name='Population': This argument renames the column containing the values from the original data to "Population" in the unpivoted DataFrame.
     
```

```python

world_data_df.drop(['Country Name'], axis = 1, inplace=True)
```

```python

world_data_df
```

```python
# Create the bar chart
plt.figure(figsize=(8, 5))  # Adjust figure size as needed
plt.bar(world_data_df['Year'], world_data_df['Population'])
plt.xlabel('Year')
plt.ylabel('Population (In Billions)')
plt.title('World Population (1960-2022)')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a faint grid for better visualization
plt.tight_layout()  # Adjust spacing between elements
# Display the chart
plt.show()
```
